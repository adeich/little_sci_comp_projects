```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
n, pA, pB = 100, 0.5, 0.6

x = np.arange(20, 80)

ax.plot(x, binom.pmf(x, n, pA), label='$A$: n=100, p=0.5')
ax.plot(x, binom.pmf(x, n, pB), label='$B$: n=100, p=0.6')

ax.legend(loc='best', frameon=False)
plt.show()
```

```python
import numpy as np
from scipy.stats import binom, norm
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
n, pA, pB = 100, 0.5, 0.6

x = np.arange(20, 80)

ax.plot(x, binom.pmf(x, n, pA), label='$A$: n=100, p=0.5')
ax.plot(x, binom.pmf(x, n, pB), label='$B$: n=100, p=0.6')
ax.plot(x, norm.pmf(x, n, pB), label='$B$: n=100, p=0.6')


ax.legend(loc='best', frameon=False)
plt.show()
```
